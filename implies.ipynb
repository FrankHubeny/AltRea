## Implies introduction and elimination

```python
from sympy.abc import P, Q
from altrea.tf import Proof
from altrea.display import show
p = Proof(P >> (Q >> P))
p.openblock(P)
p.openblock(Q)
p.reit(1)
p.closeblock()
p.implies_intro(2)
p.closeblock()
p.implies_intro(1)
show(p)
```

```python
q = eliminate_implications(p.goal)
q
```

```python
from sympy.abc import P, Q
from altrea.tf import Proof
from altrea.display import show
p2 = Proof((P | ~P) >> (P >> (Q >> P)))
show(p2)
```

```python
from sympy.abc import A, B, C
from altrea.tf import Proof
from altrea.display import show
p = Proof(A & B)
p.and_elimname = 'A E'
p.addpremise(A & B & C)
p.and_elim(1)
show(p)
```

```python
from sympy.abc import A, B
from altrea.tf import Proof
from altrea.display import show
p = Proof(B)
p.addpremise(A)
p.addpremise(A >> B)
p.implies_elim(1,2)
show(p, latex=0)
```
