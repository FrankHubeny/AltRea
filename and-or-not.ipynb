# DeMorgan Rules

1. Given $\neg(A \wedge B)$ as a premise, derive $\neg A \vee \neg B$ as the goal.
2. Given $\neg(A \vee B)$ as a premise, derive $\neg A \wedge \neg B$ as the goal.
3. Given $\neg A \vee \neg B$ as a premise, derive $\neg (A \wedge B)$ as the goal.
4. Given $\neg A \wedge \neg B$ as a premise, derive $\neg (A \vee B)$ as the goal.

```python
# DeMorgan 1 of 4: Given ~(A & B) as a premise derive the goal ~A | ~B
from sympy.abc import A, B
from altrea.tf import Proof
from altrea.display import show
p = Proof(~A | ~B, comments="DeMorgan 1 of 4")
p.addpremise(~(A & B))
p.demorgan(1)
show(p)
```

```python
# DeMorgan 2 of 4: Given ~(A | B) as a premise derive the goal ~A & ~B
from sympy.abc import A, B
from altrea.tf import Proof
from altrea.display import show
p = Proof(~A & ~B, comments="DeMorgan 2 of 4")
p.addpremise(~(A | B))
p.demorgan(1)
show(p)
```

```python
# DeMorgan 3 of 4: Given ~A | ~B as a premise derive the goal ~(A & ~B)
from sympy.abc import A, B
from altrea.tf import Proof
from altrea.display import show
p = Proof(~(A & B), comments="DeMorgan 3 of 4")
p.addpremise(~A | ~B)
p.demorgan(1)
show(p)
```

```python
# DeMorgan 4 of 4: Given ~A & ~B as a premise derive the goal ~(A | ~B)
from sympy.abc import A, B
from altrea.tf import Proof
from altrea.display import show
p = Proof(~(A | B), comments="DeMorgan 4 of 4")
p.addpremise(~A & ~B)
p.demorgan(1)
show(p)
```
