## And Elimination

```python
# Clean run with no errors

from altrea.boolean import And, Wff
from altrea.truthfunction import Proof
from altrea.display import show
A = Wff('A')
B = Wff('B')
p = Proof()
p.addgoal(A, 'one')
p.addgoal(B, 'two')
p.addpremise(And(A, B))
p.and_elim(1)
show(p, latex=1)
```
