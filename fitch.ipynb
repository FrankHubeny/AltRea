```python
# 6.10 comm conj

import pandas as pd
pd.options.display.max_colwidth=500

from altrea.boolean import Wff, Or, Not, And, Implies, Iff, Necessary, Possibly
from altrea.rules import Proof
from altrea.display import fitchnotation

commconj = Proof(name='comm conj', displayname='comm conj', description='Commutative Law for Conjunction')
p = commconj.proposition('p')
q = commconj.proposition('q')
r = commconj.proposition('r')
fitchnotation(commconj)
commconj.setlogic('fitch')
commconj.goal(Iff(And(p, q), And(q, p)))
commconj.hypothesis(And(p, q))
commconj.conjunction_elim(1, side='left')
commconj.conjunction_elim(1, side='right')
commconj.conjunction_intro(3, 2)
commconj.implication_intro()
commconj.hypothesis(And(q, p))
commconj.conjunction_elim(6, side='left')
commconj.conjunction_elim(6, side='right')
commconj.conjunction_intro(8, 7)
commconj.implication_intro()
commconj.coimplication_intro(5, 10)
commconj.displaylog()
commconj.displayproof()
```

```python
# 6.10 ass conj

from altrea.boolean import Wff, Or, Not, And, Implies, Iff, Necessary, Possibly
from altrea.rules import Proof
from altrea.display import fitchnotation

assconj = Proof(name='ass conj', displayname='ass conj', description='Associative Law for Conjunction')
p = assconj.proposition('p')
q = assconj.proposition('q')
r = assconj.proposition('r')
fitchnotation(assconj)
assconj.setlogic('fitch')
assconj.goal(Iff(And(And(p, q), r), And(p, And(q, r))))
assconj.hypothesis(And(And(p, q), r))
assconj.conjunction_elim(1, side='left')
assconj.conjunction_elim(1, side='right')
assconj.conjunction_elim(2, side='left')
assconj.conjunction_elim(2, side='right')
assconj.conjunction_intro(5, 3)
assconj.conjunction_intro(4, 6)
assconj.implication_intro()
assconj.hypothesis(And(p, And(q, r)))
assconj.conjunction_elim(9, side='left')
assconj.conjunction_elim(9, side='right')
assconj.conjunction_elim(11, side='left')
assconj.conjunction_elim(11, side='right')
assconj.conjunction_intro(10, 12)
assconj.conjunction_intro(14, 13)
assconj.implication_intro()
assconj.coimplication_intro(8, 16)
assconj.displaylog()
assconj.displayproof()
```
